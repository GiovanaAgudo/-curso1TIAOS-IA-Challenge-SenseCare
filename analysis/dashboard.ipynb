```python
import pandas as pd
import json
from pathlib import Path

# paths
PROJECT_ROOT = Path("..").resolve()
EXPORTS = PROJECT_ROOT / "analysis" / "exports"

df_events = pd.read_csv(EXPORTS / "events_flat.csv")
df_sessions = pd.read_csv(EXPORTS / "session_metrics.csv")

total_sessions = len(df_sessions)
total_events = len(df_events)
csat_mean = df_sessions["csat_mean"].mean().round(2)
dwell_mean = df_sessions["dwell_ms"].mean().round(0)
access_pct = (df_sessions["modes_enabled"]
              .apply(lambda x: len(json.loads(x)) if isinstance(x, str) else 0)
              .gt(0)
              .mean() * 100).round(1)

Markdown(f"""
### 📊 KPIs Gerais  
- **Total de sessões:** `{total_sessions}`  
- **Total de eventos:** `{total_events}`  
- **CSAT médio:** `{csat_mean}`  
- **Tempo médio de permanência (ms):** `{dwell_mean}`  
- **Sessões com modos de acessibilidade:** `{access_pct}%`  
""")
```

```python
import matplotlib.pyplot as plt

def plot_channel_distribution():
    plt.figure(figsize=(6,4))
    df_sessions["channel_main"].value_counts().plot(kind="bar")
    plt.title("Distribuição de Sessões por Canal Principal")
    plt.xlabel("Canal")
    plt.ylabel("Número de sessões")
    plt.show()

plot_channel_distribution()
```
